```python
#Sitios web de la extraccion de datos
url ="http://books.toscrape.com/"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
#recorriendo los datos y almacenandolos en una lista
books = soup.find_all("article", class_="product_pod")
book_list = [] #crear una lista vacia
for book in books:#creando estructura repetitiva for
  title = book.h3.a['title']
  price = book.find("p", class_="price_color").text
  book_list.append({"Title": title, "Price": price})  #append (agregar valores a la lista )
```

```python
df_books = pd.DataFrame(book_list)
df_books
```

```python
#convertir todos los valores de la columna Price a dolares
exchange_rate = 1.27
#Elimina el simbolo de la moneda y convierte a float
df_books['Price'] = df_books['Price'].astype(str).str.replace('£', '').astype(float)

#convertir a dolares
df_books['Price(USD)'] = df_books['Price'] * exchange_rate
df_books
```

```python
df_books_sel =df_books[['Title', 'Price(USD)']]
df_books_sel
```
